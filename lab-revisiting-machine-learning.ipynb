```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
pd.set_option('display.max_columns', None)
```

```python
data = pd.read_csv('learningSet.csv')
```

```python
categorical = data.select_dtypes([object])
categorical.head()
```

```python
categorical.isna().sum()
```

```python
def column_null(categorical, threshold=0.85):
    categorical = categorical.copy()
    nulls_percent_df = pd.DataFrame(categorical.isna().sum()/len(categorical)).reset_index()
    nulls_percent_df.columns = ['column_name', 'nulls_percentage']
    drop_columns_list = nulls_percent_df[nulls_percent_df['nulls_percentage']>threshold]
    return list(drop_columns_list['column_name'])
```

```python
drop_list = column_null(categorical, 0.85)
drop_list
```

```python
drop_list.append('OSOURCE')
drop_list.append('ZIP')
drop_list
```

```python
print(categorical['GENDER'].value_counts())
categorical['GENDER'] = categorical['GENDER'].fillna('F')
```

```python
categorical['GENDER'] = categorical['GENDER'].apply(lambda x: x.replace(" ", "other"))
categorical['GENDER'] = categorical['GENDER'].apply(lambda x: x.replace("U", "other"))
categorical['GENDER'] = categorical['GENDER'].apply(lambda x: x.replace("J", "other"))
categorical['GENDER'] = categorical['GENDER'].apply(lambda x: x.replace("C", "other"))
categorical['GENDER'] = categorical['GENDER'].apply(lambda x: x.replace("A", "other"))
print(categorical['GENDER'].value_counts())
```
